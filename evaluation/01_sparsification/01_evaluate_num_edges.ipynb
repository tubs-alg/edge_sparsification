# Getting some general information on the sparsified graphs

```python
import pandas as pd
import seaborn as sns
sns.set_theme()
import matplotlib.pyplot as plt
sns.set_theme()
plt.rcParams.update({'figure.max_open_warning': 0
                    ,"text.usetex": True
                    ,"font.family": "serif"
                    })
sns.set_palette("tab10")
PAPER_WIDTH = 10
```

```python
data = pd.read_json("./00_sparsification_data.json.zip")
```

```python
data
```

```python
data["sparsification_technique"].unique()
```

```python
angdist = ["complete", "delaunayT", "wspd2-1", "yao6-2"]
angtsp = ["complete", "onion-0", "onion-1-1", "wspd7-3", "yao6-2",  "wspd2-1", "wspd2-3",]
maxarea = ["complete", "theta12-2", "wspd2-3", "yao12-2"]
minarea= ["complete", "theta12-2", "wspd2-3", "yao12-2"]
mpp =["complete", "delaunayT", "greedyT", "mwT", "yao6-2"]
selection = list(set(angdist + angtsp +maxarea+minarea+mpp))
#selection = ['complete', 'mwT', 'greedyT', 'onion-0','wspd2-1' , 'yao6-2', ]
sorted(selection)
```

```python
t = data[data["sparsification_technique"].isin(selection)].copy()
t["sparsification_technique"].replace({'delaunayT': "triangulation",
 'greedyT': "triangulation",
 'mwT': "triangulation"}, inplace=True)
t["sparsification_technique"].replace({'theta12-2': "cone(12)",
 'yao12-2': "cone(12)",
 'yao6-2': "cone(6)"}, inplace=True)
t["num_vertices"] = t["num_vertices"].apply(lambda x: round(x/10)*10)
t=t[t["num_vertices"]<=500]
t = t.groupby(["num_vertices", "sparsification_technique"])[["num_edges"]].mean().reset_index()
```

```python
order =  {  'complete': 0,
 'triangulation': 1,
 'onion-0': 2,
 'onion-1-1': 3,
 'cone(12)': 5,
 'wspd2-1': 6,
 'wspd2-3': 7,
 'wspd7-3': 8,
 'cone(6)' : 4,
 }
t["order"]=t["sparsification_technique"].apply(lambda x: order[x])
t.sort_values(by=["order","num_vertices"], inplace=True)
```

```python
techniques = t["sparsification_technique"].unique()
palette = sns.color_palette("tab10")
colors = {
   'complete': palette[0],
 'triangulation': palette[1],
 'onion-0': palette[2],
 'onion-1-1': palette[2],
 'cone(12)': palette[3],
 'wspd2-1': palette[4],
 'wspd2-3': palette[4],
 'wspd7-3': palette[4],
 'cone(6)' : palette[3],
}
palette = sns.color_palette([colors[t] for t in techniques])
line_styles = ['-', ":", "--", "-."]
styles = {
       'complete': line_styles[0],
 'triangulation': line_styles[0],
 'onion-0': line_styles[0],
 'onion-1-1': line_styles[1],
 'cone(12)': line_styles[1],
 'wspd2-1': line_styles[0],
 'wspd2-3': line_styles[1],
 'wspd7-3': line_styles[2],
 'cone(6)' : line_styles[0],
}
line_styles = [styles[t] for t in techniques]
```

```python
plt.figure(figsize=(0.6*PAPER_WIDTH, 4))
t["technique"]=t["sparsification_technique"]

ax = sns.lineplot(data=t, x="num_vertices", y="num_edges", hue="technique", style="technique", palette=palette)
for i, l in enumerate(ax.lines):
    l.set_linestyle(line_styles[i%len(line_styles)])
ax.legend(title=None)
plt.xlabel("Number of vertices")
plt.ylabel("Number of edges")
plt.tight_layout()
plt.savefig("output/num_edges_lineplot.pdf")
plt.show()
```

```python
plt.figure(figsize=(0.4*PAPER_WIDTH, 4))
ns = [20, 100, 400]
t_ = t[t["num_vertices"].isin(ns)].copy()

t_.rename(columns={"num_vertices": "n", "num_edges": "m"}, inplace=True)
t_["complete"]=0
for n in ns:
    t_.loc[t_["n"]==n, "complete"]=t_[t_["n"]==n]["m"].max()
t_ = t_[t_["technique"]!="complete"].copy()
t_["x"]=100*t_["m"]/t_["complete"]
t_ = t_.groupby(["n", "technique"])[["x", "order"]].mean().reset_index()
t_.sort_values(by="order", inplace=True)
sns.barplot(data=t_, y="technique", x="x", hue="n")
plt.legend(title="Vertices")
plt.xlabel("Percentage of edges (\%)")
plt.ylabel("Sparsification")
plt.xlim(0, 100)
plt.tight_layout()
plt.savefig("output/num_edges_barplot.pdf")
plt.show()
```
