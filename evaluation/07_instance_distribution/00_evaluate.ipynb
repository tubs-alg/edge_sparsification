```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
plt.rcParams.update({'figure.max_open_warning': 0
                    ,"text.usetex": True
                    ,"font.family": "serif"
                    })
sns.set_palette("tab10")
PAPER_WIDTH = 10
```

```python
def plot(t):
    t = t.copy()
    t["type"].replace({"uniform_real": "uniform", "uniform_int": "uniform", "TSPLIB_subset": "TSPLIB", "image_edge": "image", "image_brightness": "image"}, inplace=True)
    #t["num_vertices"] = t["num_vertices"].apply(lambda x: round(x/5)*5)
    sns.histplot(data=t, x="num_vertices", hue="type", multiple="dodge", shrink=.8)
    plt.show()
```

```python
# Gather instance types
instance_types = pd.read_json("../01_sparsification/00_sparsification_data.json.zip")[["filename_y", "type", "num_vertices"]].drop_duplicates(subset=["filename_y", "type", "num_vertices"]).rename(columns={"filename_y": "instance"}).set_index("instance")
instance_types
```

```python
# angular tsp
t = pd.read_json("../02_angular_tsp/00_angular_data.json.zip")
t["instance"] = t["filename_instance"]#.apply(lambda x: x.split(".")[0])
instances = t["instance"].unique().tolist()
angular_tsp_instances = instance_types.loc[instance_types.index.isin(instances)]
angular_tsp_instances
```

```python
plot(angular_tsp_instances)
```
